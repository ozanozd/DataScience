```python
import networkx as nx

nodes = []

with open("SMS-network.txt") as f: #reading file
    for line in f: #line by line
        nodes.append((int(line.split(' ')[0]), int(line.split(' ')[1])))
               
#print(nodes)
```

```python
# Networkx graph
graph = nx.Graph() 
 
# We adding the edges to the graph
# One direction or two direction should be checked
graph.add_edges_from(nodes)
```

```python
# Draw the resulting graph
nx.draw_networkx(graph, with_labels=True, font_weight='bold')
```
